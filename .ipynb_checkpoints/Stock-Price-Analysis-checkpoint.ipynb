## 1) Data Collection

```python
company_list = [
    'D:\\Udemy Projects Data analysis\\Stock-Market-Data-Analysis\\individual_stocks_5yr_data\\AAPL_data.csv',
    'D:\\Udemy Projects Data analysis\\Stock-Market-Data-Analysis\\individual_stocks_5yr_data\\AMZN_data.csv',
    'D:\\Udemy Projects Data analysis\\Stock-Market-Data-Analysis\\individual_stocks_5yr_data\\GOOG_data.csv',
    'D:\\Udemy Projects Data analysis\\Stock-Market-Data-Analysis\\individual_stocks_5yr_data\\MSFT_data.csv'
]
```

```python
df = pd.DataFrame()

for file in company_list:
    current_df = pd.read_csv(file)
    df = pd.concat([df, current_df], ignore_index=True)
```

## 2) Analysing change in price of the stock overtime !

```python
df.info()
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df.dtypes
```

```python
stock_name = df["Name"].unique()
```

```python
df["year"] = df["date"].dt.year
```

```python
df.year.unique()
```

```python
plt.figure(figsize = (20,12))

for index, company in enumerate(stock_name, 1):
    plt.subplot(2, 2, index)   # creating subplot for each stock
    filter = df["Name"] == company
    data = df[filter]
    plt.plot(data["date"], data["close"]) # plotting 'date' vs 'close'(i.e the closing price of the stock for a particular date)
    plt.title(company)
```

#### As we can see that `Amazon`, `Google`, `Microsoft` has a `Trending line` i.e in `each year` the stock price is `increasing to an extent`.
#### whereas in case of `Apple` there is a `peak` in `2015`, then a `fall` in `2016` and again increasing with a `peak` in `2018`.

## 3) Analysing moving average of various stocks

```python
df.head()
```

```python
new_data = df.drop(columns = "year")        # We dropped year column because we don't need it for this problem
```

```python
new_data.head()
```

```python
new_data["close"].rolling(window=7).mean().head(10)
```

```python
new_data_ma = new_data.copy()
```

```python
ma_value = [10, 20, 50]

for ma in ma_value:
    new_data_ma["close_" + str(ma)] = new_data_ma["close"].rolling(ma).mean()
```

```python
# new_data_ma.head(20)
```

```python
new_data_ma.set_index("date", inplace = True)
```

#### We set the `index` of the data to `date` feature

```python
# new_data_ma.head(5)
```

```python
new_data_ma.columns
```

```python
plt.figure(figsize = (20,13))

for index, company in enumerate(stock_name, 1):
    plt.subplot(2, 2, index)
    filter = new_data_ma["Name"] == company
    data = new_data_ma[filter]
    data[['close_10', 'close_20', 'close_50']].plot(ax = plt.gca())
    plt.title(company)
```

#### Every stock seems similar to the previous graphs, but `MSFT`, the graph moving average is probably same after `2013`.

## 6) Performing analysis to see if stocks `['apple', 'google', 'amazon', 'microsoft']` has any co-relation between the closing prices of each

```python
company_list
```

```python
apple = pd.read_csv(company_list[0])
amzn = pd.read_csv(company_list[1])
google = pd.read_csv(company_list[2])
msft = pd.read_csv(company_list[3])
```

```python
company_closing_prices = pd.DataFrame()
```

```python
company_closing_prices["apple_close"] = apple["close"]
company_closing_prices["amzn_close"] = amzn["close"]
company_closing_prices["goog_close"] = google["close"]
company_closing_prices["msft_close"] = msft["close"]
```

```python
company_closing_prices.head()
```

```python
sns.pairplot(company_closing_prices)
```

#### We can see that `amazon` and `microsoft` has a `higher co-relation` between each other.
#### Let's check using `heatmap`.

```python
sns.heatmap(company_closing_prices.corr(), annot=True)
```

#### So we can see a higher co-relation between the above stocks mentioned.
#### Therefore, we can conclude that when `closing price of [Amazon] increases, closing price of [Microsoft] may increase` based on the data.
